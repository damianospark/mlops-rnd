# Train and deploy a model using features

```python
training_df.predict_prob_group.describe
```

```python
import statsmodels.api as sm
```

```python
x = training_df.predict_prob
X = sm.add_constant(x)
```

```python
y = training_df.predict_prob_group
```

```python
model = sm.OLS(y.astype(float), X.astype(float))
res = model.fit()
```
